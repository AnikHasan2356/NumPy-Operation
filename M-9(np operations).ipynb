```python
import numpy as np
arr=np.random.randint(1,10,size=(5,10))
b=arr.reshape(2,25)
print(arr)
print("b=",b)
print("\n")
#Flattening or convert into 1D array:

c=np.ravel(arr,order='F')         #ordering according to row
print("C=",c)
print(" \n")
d=np.ravel(arr,order='c')         #ordering according to row
print("D=",d)
```

```python
#Concatenet:
a=np.random.randint(1,5,size=(2,3))
b=np.random.randint(1,5,size=(2,3))
print("a=",a,"\n","\n","b=",b,"\n")

row=np.concatenate((a,b),axis=0) #for row concatenate col should be same in deimension of those matrix
print(row,"\n")
row=np.concatenate((a,b),axis=1) #for col concatenate row should be same in deimension of those matrix
print(row,"\n")
```

```python
#Transpose matrix:
a=np.random.randint(1,5,size=(2,3))
mat=a.T
print(mat,"\n")


#array_split and split: 
arr=np.random.randint(1,10,size=(10,))     #when data equally division isn't possible
b=np.array_split(arr,3)
print("b=",b,"\n")

arr1=np.random.randint(1,10,size=(9,))     #when data equally division possible
c=np.split(arr1,3)
print("c=",c)
```

```python
#vectorization is faster in array:
a=np.random.randint(1,5,size=(2,3))
b=np.random.randint(1,5,size=(2,3))
print(a,"\n",b,"\n")
add=a+b
print(add)
  #same as:
sub=a-b
mul=a*b
div=a/b
print("mul=",mul,"\n")
#Trigonometric: sin, cos, ... values in rediant
sine=np.sin(a)
print(sine)

  #degree to rediant and vice versa:
rad=np.deg2rad(a)
print(rad)  #same for rad2deg

#log: how many times a division is possible by the base of log
log=np.log10(a)
log1=np.log2(b)
sqrt=np.sqrt(a)
total_sum=sum(a)
cummulative_sum=np.cumsum(a)
```

```python
#Logical operator:>,<,==,..
a=np.random.randint(1,5,size=(2,3))
b=np.random.randint(1,5,size=(2,3))
greater=a>b
print(greater)

#To check true and false after complete a operation: all(), any()
check1=np.all(greater)
print(check1)
check2=np.any(greater)
print(check2)
```

```python
#Sorting 

x = np.array([2, 4, 1, 9, 4])

z = x.copy()
print(z)

z.sort()   #Inplace Sorting
print(z)
print(x)
print("\n")

a=np.array([5,3,4,7,1])
y=np.sort(a)  #copy sorting
print(y)

arr2 =np.array([[1,4,3,2,5],
                [6,9,8,7,10]])
row_sort=np.sort(arr2,axis=1)
print(row_sort)
```

```python
#Searching:
arr2 =np.array([[1,4,3,2,5],
                [6,9,8,7,10]])
index=np.where(x>arr2)           #finding array position/index
value=np.where(x>arr2,x,10)      #cand modify the array
print(index)
print(value)

max=np.argmax(arr2)  #finding max value's index
min=np.argmin(arr2)  #finding max value's index
print(max)
print(min)

#Count:
count=np.count_nonzero(arr2>10)
print(count)
unique=np.unique(arr2,return_counts=True)
print(unique)
```

```python
#correlation:
a=np.random.randint(1,5,size=(2,3))
b=np.random.randint(1,5,size=(2,3))
print("a=",a,"\n","\n","b=",b,"\n")

row=np.concatenate((a,b),axis=0) #for row concatenate col should be same in deimension of those matrix

row=np.concatenate((a,b),axis=1) #for col concatenate row should be same in deimension of those matrix

correltn=np.corrcoef(row)
print(correltn)
```
